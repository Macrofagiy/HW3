```python
import numpy as np
import random
import matplotlib.pyplot as plt
import math
```

```python
x = np.linspace(5, 10, 11)


fig, ax = plt.subplots()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
y = x*50
z = x*0 + 900
ax.plot(x, y, linestyle='solid')
plt.xlabel('Accelereting voltage [kV]')
plt.ylabel('Current nA')
plt.title('(a)')
plt.plot(x, y,  marker = 's', label='line 1', color = 'black', markerfacecolor = 'black',
        markeredgecolor = 'black')
plt.plot(x, z,  marker = 'o', label='line 1', color = 'black', markerfacecolor = 'white',
        markeredgecolor = 'black')
ax.set_xlim([5, 10])
ax.set_ylim([1, 1000])

fig, bx = plt.subplots()
x = np.linspace(0, 1000, 300)
y = [random.uniform(3.75, 4.25) for i in x]
plt.xlabel('Time [8]')
plt.ylabel('Sample current nA')
plt.title('(b)')
plt.plot(x, y, label='line 1', color = 'black')
bx.set_ylim([0, 10])
bx.spines['top'].set_visible(False)
bx.spines['right'].set_visible(False)


plt.figure(figsize=(7,4))

px = plt.subplot(121)
import matplotlib.image as mpimg
img = mpimg.imread('imgg.png')
plt.imshow(img)
px.spines['top'].set_visible(False)
px.spines['right'].set_visible(False)
px.spines['bottom'].set_visible(False)
px.spines['left'].set_visible(False)
px.tick_params(axis='x', colors='white')
px.tick_params(axis='y', colors='white')

cx = plt.subplot(122)
plt.xlabel("xlabel")
plt.ylabel("ylabel")
x = np.linspace(0, 100, 5)
y = [random.uniform(0, 10) for i in x]
plt.title('(с)')
plt.plot(x, y, label='line 1', color = 'black')
cx.set_ylim([0, 10])
cx.spines['top'].set_visible(False)
cx.spines['right'].set_visible(False)
```
